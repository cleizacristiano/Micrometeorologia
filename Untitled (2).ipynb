```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
from datetime import datetime
from pathlib import Path
from matplotlib.dates import DateFormatter
import glob

# Diretórios
alta_freq_dir = Path("C:/WPy64-31241/notebooks/micro/dados/alta")
baixa_freq_dir = Path("C:/WPy64-31241/notebooks/micro/dados/baixa")

# Funções auxiliares
def clean_data(df):
    df.replace(-9999, np.nan, inplace=True)
    return df.dropna()

def rotacao_2d(u, v, w):
    theta = np.arctan2(np.nanmean(v), np.nanmean(u))
    u2 = u * np.cos(theta) + v * np.sin(theta)
    v2 = -u * np.sin(theta) + v * np.cos(theta)
    return u2, v2, w

def fluxo_co2_corrigido(w, c, q, T):
    ma, mv, mc, rho_ar = 28.96, 18.0, 44.0, 1.2
    try:
        Fc = np.cov(w, c)[0, 1]
        webb1 = (ma/mv) * (np.nanmean(c) * mc * 1e-6 / rho_ar) * np.cov(w, q)[0, 1]
        webb2 = (1 + (ma/mv) * (np.nanmean(q) * mv * 1e-6 / rho_ar)) * \
                 (np.nanmean(c) / (np.nanmean(T) + 273.15)) * np.cov(w, T)[0, 1]
        return Fc * 1000, (Fc + webb1 + webb2) * 1000
    except:
        return np.nan, np.nan

def calc_ect_ustar(u, v, w):
    u_prime = u - np.nanmean(u)
    v_prime = v - np.nanmean(v)
    w_prime = w - np.nanmean(w)
    ustar = np.sqrt(np.sqrt((np.nanmean(u_prime * w_prime) ** 2 +
                             np.nanmean(v_prime * w_prime) ** 2)))
    ect = 0.5 * (np.nanvar(u_prime) + np.nanvar(v_prime) + np.nanvar(w_prime))
    return ect, ustar

def carregar_dados_baixa():
    arquivos_111 = glob.glob(str(baixa_freq_dir / "FA_111_*.dat"))
    arquivos_113 = glob.glob(str(baixa_freq_dir / "FA_113_*.dat"))
    df_111 = pd.concat([pd.read_csv(f, header=None) for f in arquivos_111])
    df_113 = pd.concat([pd.read_csv(f, header=None) for f in arquivos_113])

    df_111.columns = ["timestamp", "tmp1", "tmp2", "SWin", "SWout", "LWin", "LWout", "tmp8", "tmp9", "tmp10", "tmp11", "tmp12"]
    df_113.columns = ["timestamp", "tmp1", "tmp2", "tmp3", "tmp4", "tmp5", "tmp6", "SHF1", "SHF2", "SHF3", "precip"]

    df_111['timestamp'] = pd.to_datetime(df_111['timestamp'])
    df_113['timestamp'] = pd.to_datetime(df_113['timestamp'])
    df_113['precip'] = df_113['precip'].clip(lower=0)

    return df_111, df_113

def processar_dados_alta():
    arquivos_110 = glob.glob(str(alta_freq_dir / "FA_110_*.dat"))
    resultados = []
    for arquivo in arquivos_110:
        df = pd.read_csv(arquivo, header=None)
        if df.shape[1] < 8: continue
        df = clean_data(df)
        if df.shape[0] < 10: continue

        timestamp = pd.to_datetime(df.iloc[0, 0])
        u, v, w = df[1], df[2], df[3]
        T, CO2, H2O = df[4], df[6], df[7]

        u2, v2, w2 = rotacao_2d(u, v, w)
        Fc, Fcw = fluxo_co2_corrigido(w2, CO2, H2O, T)

        LE = 18 * 2.45 * np.cov(w2, H2O)[0, 1] if H2O.notna().all() else np.nan
        H = 1005 * np.cov(w2, T)[0, 1] if T.notna().all() else np.nan
        ect, ustar = calc_ect_ustar(u2, v2, w2)

        resultados.append({
            "timestamp": timestamp,
            "Fc": Fc,
            "Fcw": Fcw,
            "H": H,
            "LE": LE,
            "ECT": ect,
            "Ustar": ustar
        })

    return pd.DataFrame(resultados)

alta_df = processar_dados_alta()
baixa_111, baixa_113 = carregar_dados_baixa()

alta_df['timestamp_30min'] = alta_df['timestamp'].dt.floor('30min')
baixa_111['timestamp_30min'] = baixa_111['timestamp'].dt.floor('30min')
baixa_113['timestamp_30min'] = baixa_113['timestamp'].dt.floor('30min')

media_rad = baixa_111.groupby('timestamp_30min')[['SWin', 'SWout', 'LWin', 'LWout']].mean()
media_shf = baixa_113.groupby('timestamp_30min')[['SHF1', 'SHF2', 'SHF3']].mean()

total = alta_df.merge(media_rad, on='timestamp_30min', how='left')
total = total.merge(media_shf, on='timestamp_30min', how='left')

total['Rn'] = (total['SWin'] - total['SWout']) + (total['LWin'] - total['LWout'])
total['G'] = total[['SHF1', 'SHF2', 'SHF3']].mean(axis=1)
total['Fechamento'] = total['H'] + total['LE'] + total['G'] - total['Rn']

total.to_csv("resultados_fluxos.csv", index=False)

total['hora'] = total['timestamp_30min'].dt.hour + total['timestamp_30min'].dt.minute / 60
media_diaria = total.groupby('hora')[['Fc', 'Fcw', 'H', 'LE', 'G', 'Rn', 'ECT', 'Ustar']].mean()
```

```python
# Gráficos de Fluxo CO2 (mg/m²/s)
plt.figure()
total.plot(x='timestamp_30min', y=['Fc', 'Fcw'], title='Fluxo de CO2')
plt.ylabel("Fluxo CO2 (mg/m²/s)")
plt.savefig("fluxos_co2.png", bbox_inches='tight')
```

```python
# Gráficos de Energia Cinética Turbulenta
plt.figure()
total.plot(x='timestamp_30min', y='ECT', title='Energia Cinética Turbulenta')
plt.ylabel("ECT (J/kg)")
plt.savefig("ect.png", bbox_inches='tight')
```

```python
# Gráficos de Velocidade de Fricção
plt.figure()
total.plot(x='timestamp_30min', y='Ustar', title='Velocidade de Fricção')
plt.ylabel("u* (m/s)")
plt.savefig("ustar.png", bbox_inches='tight')
```

```python
# Gráficos de Ciclo Médio Diário de CO2
plt.figure()
media_diaria[['Fc', 'Fcw']].plot(title='Ciclo Médio Diário de CO2')
plt.savefig("ciclo_diario_co2.png", bbox_inches='tight')
```

## com filtro de -50 ≤ LE ≤ 50
Os valores fora do intervalo não são considerados de forma alguma — não são zerados, nem substituídos por NaN, apenas excluídos do DataFrame usado na análise.

```python
# Gráfico composto: variáveis do balanço de energia em subplots separados com filtro de LE entre -50 e 50
fig, axs = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

axs[0].plot(total['timestamp_30min'], total['H'], color='red')
axs[0].set_ylabel("H (W/m²)")
axs[0].set_title("Fluxo Sensível (H)")

mascara_le = (total['LE'] >= -50) & (total['LE'] <= 50)
axs[1].plot(total.loc[mascara_le, 'timestamp_30min'], total.loc[mascara_le, 'LE'], color='blue')
axs[1].set_ylabel("LE (W/m²)")
axs[1].set_title("Fluxo Latente (LE) (Filtrado -50 a 50)")

axs[2].plot(total['timestamp_30min'], total['G'], color='green')
axs[2].set_ylabel("G (W/m²)")
axs[2].set_title("Fluxo de Calor no Solo (G)")

axs[3].plot(total['timestamp_30min'], total['Rn'], color='orange')
axs[3].set_ylabel("Rn (W/m²)")
axs[3].set_title("Saldo de Radiação (Rn)")
axs[3].set_xlabel("Tempo")

plt.tight_layout()
plt.savefig("balanco_energia_subplots_50.png", bbox_inches='tight')
```

```python
# Gráfico composto: ciclo médio diário das variáveis do balanço de energia com filtro de LE entre -50 e 50
fig, axs = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

axs[0].plot(media_diaria.index, media_diaria['H'], color='red')
axs[0].set_ylabel("H (W/m²)")
axs[0].set_title("Ciclo Médio Diário - Fluxo Sensível (H)")

media_diaria_LE_filtrada = total[(total['LE'] >= -50) & (total['LE'] <= 50)].groupby('hora')['LE'].mean()
axs[1].plot(media_diaria_LE_filtrada.index, media_diaria_LE_filtrada, color='blue')
axs[1].set_ylabel("LE (W/m²)")
axs[1].set_title("Ciclo Médio Diário - Fluxo Latente (LE) (Filtrado -50 a 50)")

axs[2].plot(media_diaria.index, media_diaria['G'], color='green')
axs[2].set_ylabel("G (W/m²)")
axs[2].set_title("Ciclo Médio Diário - Fluxo de Calor no Solo (G)")

axs[3].plot(media_diaria.index, media_diaria['Rn'], color='orange')
axs[3].set_ylabel("Rn (W/m²)")
axs[3].set_title("Ciclo Médio Diário - Saldo de Radiação (Rn)")
axs[3].set_xlabel("Hora do dia")

plt.tight_layout()
plt.savefig("ciclo_diario_balanco_subplots_50.png", bbox_inches='tight')
```

```python
# Fechamento do balanço de energia com LE entre -50 e 50
total_filtrado = total[(total['LE'] >= -50) & (total['LE'] <= 50)]
x_var = total_filtrado['Rn'] - total_filtrado['G']
y_var = total_filtrado['H'] + total_filtrado['LE']

plt.figure()
plt.scatter(x_var, y_var, alpha=0.6, color='black', s=10)
plt.plot(x_var, x_var, linestyle='--', color='blue')
sns.regplot(x=x_var, y=y_var, scatter=False, color='limegreen')
plt.xlabel("Rn - G (W/m²)")
plt.ylabel("H + LE (W/m²)")
plt.title("Fechamento do Balanço de Energia (LE entre -50 e 50)")
plt.xlim(-100, 800)
plt.ylim(-100, 800)
plt.savefig("fechamento_balanco_50.png", bbox_inches='tight')
```
